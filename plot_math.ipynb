```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import math

def interactive_plot(ts, C):
    M = 0.05 # 50g
    L = 0.5 #1 meter
    g = 9.81 # speed of gravity
    
    t = np.linspace(0,1,30)

    theta = ts * np.cos(math.sqrt((C + (M*L*g))/(M*(L**2))) * t)

    # Force the axis to be constnat (no dynamic adjusting)
    plt.figure(1)
    plt.clf()
    plt.plot(t, theta)
    plt.xlabel('t-axis')
    plt.ylabel('θ-axis')
    plt.title('θ = f(t)')
    plt.grid(True)

    plt.xlim(0,1)
    plt.ylim(-10,10)

    plt.show()

interact(interactive_plot, ts=(-10, 10, 1), C=(0, 0.5, 0.1))
```
