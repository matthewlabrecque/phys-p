```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, IntSlider, interact
import math

def interactive_plot(θᵢ, κ):
    M = 0.05 # 50g
    L = 0.5 # 1/2 meter
    g = 9.81 # gravitational acceleration
    
    t = np.linspace(0,1.5,300)

    theta = θᵢ * np.cos(math.sqrt((κ + (M*L*g))/(M*(L**2))) * t)

    # Force the axis to be constant (no dynamic adjusting)
    plt.figure(1)
    plt.clf()
    plt.plot(t, theta)
    plt.xlabel('t-axis')
    plt.ylabel('θ-axis')
    plt.title('θ = f(t)')
    plt.grid(True)

    plt.xlim(0,1.5)
    plt.ylim(-10,10)

    plt.show()
theta_slider = IntSlider(value=5, min=-10, max=10, step=0.5, description='θᵢ')
kappa_slider = FloatSlider(value=0.3, min=0, max=0.5, step=0.05, description='κ')
interact(interactive_plot, θᵢ=theta_slider, κ=kappa_slider)
```
